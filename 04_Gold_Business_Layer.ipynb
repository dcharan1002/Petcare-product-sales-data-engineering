```python
from pyspark.sql.functions import *

silver_df = spark.table("workspace.default.silver_pet_store_records")

display(silver_df)
```

Gold Table 1 - Product Performance


```python
gold_product = (
    silver_df
    .groupBy("product_category")
    .agg(
        sum("sales").alias("total_sales"),
        sum("estimated_revenue").alias("total_revenue"),
        avg("rating").alias("average_rating"),
        count("*").alias("number_of_products")
    )
)
```

```python
gold_product.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.gold_product_performance")
```

```python
display(
spark.table("workspace.default.gold_product_performance")
)
```

Gold Table 2 — Vendor Performance


```python
gold_vendor = (
    silver_df
    .groupBy("vendor_id")
    .agg(
        sum("estimated_revenue").alias("vendor_revenue"),
        avg("rating").alias("average_rating"),
        count("*").alias("products_supplied")
    )
)
```

```python
gold_vendor.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.gold_vendor_performance")
```

Gold Table 3 — Country Sales

```python
gold_country = (
    silver_df
    .groupBy("country")
    .agg(
        sum("estimated_revenue").alias("country_revenue"),
        sum("sales").alias("total_units_sold")
    )
)
```

```python
gold_country.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.gold_country_sales")
```

Gold Table 4 — Pet Type Analytics

```python
gold_pet = (
    silver_df
    .groupBy("pet_type")
    .agg(
        sum("estimated_revenue").alias("revenue"),
        avg("rating").alias("average_rating"),
        count("*").alias("products")
    )
)
```

```python
gold_pet.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.gold_pet_type")
```

Gold Table 5 — Customer Re-buy Analysis

```python
gold_rebuy = (
    silver_df
    .groupBy("re_buy")
    .agg(
        sum("estimated_revenue").alias("revenue"),
        avg("rating").alias("average_rating"),
        count("*").alias("transactions")
    )
)
```

```python
gold_rebuy.write \
.format("delta") \
.mode("overwrite") \
.saveAsTable("workspace.default.gold_rebuy")
```
