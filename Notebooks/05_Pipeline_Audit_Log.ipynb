```python
from pyspark.sql.functions import current_timestamp, lit
```

```python
audit_data = [
    ("petcare_product_pipeline", "pet_store_records_2020", "bronze_pet_store_records", spark.table("workspace.default.bronze_pet_store_records").count(), "SUCCESS"),
    ("petcare_product_pipeline", "bronze_pet_store_records", "silver_pet_store_records", spark.table("workspace.default.silver_pet_store_records").count(), "SUCCESS"),
    ("petcare_product_pipeline", "silver_pet_store_records", "gold_product_performance", spark.table("workspace.default.gold_product_performance").count(), "SUCCESS"),
    ("petcare_product_pipeline", "silver_pet_store_records", "gold_vendor_performance", spark.table("workspace.default.gold_vendor_performance").count(), "SUCCESS"),
    ("petcare_product_pipeline", "silver_pet_store_records", "gold_country_sales", spark.table("workspace.default.gold_country_sales").count(), "SUCCESS"),
    ("petcare_product_pipeline", "silver_pet_store_records", "gold_pet_type", spark.table("workspace.default.gold_pet_type").count(), "SUCCESS"),
    ("petcare_product_pipeline", "silver_pet_store_records", "gold_rebuy", spark.table("workspace.default.gold_rebuy").count(), "SUCCESS")
]

audit_df = spark.createDataFrame(
    audit_data,
    ["pipeline_name", "source_table", "target_table", "row_count", "status"]
)

audit_df = audit_df.withColumn("run_timestamp", current_timestamp())

display(audit_df)
```

```python
audit_df.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("workspace.default.pipeline_audit_log")
```
