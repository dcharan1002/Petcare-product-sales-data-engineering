```python
raw_df = spark.table("workspace.default.pet_store_records_2020")

display(raw_df)
```

```python
print("Total rows:", raw_df.count())
print("Total columns:", len(raw_df.columns))

raw_df.printSchema()
```

```python
from pyspark.sql.functions import current_timestamp, lit

bronze_df = (
    raw_df
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_system", lit("kaggle_pet_store_records_2020"))
    .withColumn("pipeline_layer", lit("bronze"))
)
```

```python
bronze_df.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.default.bronze_pet_store_records")
```
